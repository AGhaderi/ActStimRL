```python
import numpy as np
import pandas as pd
```

```python
# read collected data across data
behAll = pd.read_csv('/mnt/projects/7TPD/bids/derivatives/fMRI_DA/data_BehModel/originalfMRIbehFiles/AllBehData/behAll.csv')
behAll_columns = behAll.rename(columns={'sub_ID':'subject', 'session':'time', 'block':'condition', 'patient':'diagnosis', 'correctChoice':'rewarded'})
behAll_columns.to_csv('/mnt/projects/7TPD/bids/derivatives/fMRI_DA/data_BehModel/originalfMRIbehFiles/AllBehData/behAll_columns.csv', 
                      columns={'subject', 'time', 'condition', 'diagnosis', 'rewarded'}, index=False)
```

```python
behAll_proportion = behAll.groupby(['sub_ID', 'session', 'block', 'patient'], as_index=False)['correctChoice'].mean()
behAll_proportion = behAll_proportion.rename(columns={'sub_ID':'subject', 'session':'time', 'block':'condition', 'patient':'diagnosis', 'correctChoice':'proportion_rewarded'})
behAll_proportion.to_csv('/mnt/projects/7TPD/bids/derivatives/fMRI_DA/data_BehModel/originalfMRIbehFiles/AllBehData/behAll_proportion.csv', index=False)
```
