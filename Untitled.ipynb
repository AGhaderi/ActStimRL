```python
import numpy as np
import pandas as pd
import stan
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az
```

```python
# reading the RL model span
file_name = 'models/RL_W.stan' 
file_read = open(file_name, 'r')
stan_model = file_read.read()

# reading the observed data
examples_dir = "../data/sub-X99071_ses-01_task-DA_run-2_beh_20210422_14-27-10.csv"
df = pd.read_csv(examples_dir)
dfAct = df[df.block == 'Stim']
subName = 'sub X99071'
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
resAct = dfAct['pushed'].to_numpy().astype(int)
pushed = np.where(resAct==1)[0]
pulled = np.where(resAct==0)[0]
noRes  = np.where(resAct < 0)

corrAct= dfAct['pushCorrect']
pushCorr = np.where(corrAct==1)[0]
pulledCorr = np.where(corrAct==0)[0]


pltData = ({'label': ['pushed', 'pushCorrect', 'noRes', 'pullCorrect', 'pulled'],
            'choise': [pushed, pushCorr, noRes, pulled, pulledCorr]})
dfPlot = pd.DataFrame(pltData)
```

```python
fig = plt.figure(figsize=(10, 6), tight_layout=True)

columns = 2
rows = 2

y = [3.2 ,3, 2.1 , 1.2, 1] 
markers = ['v', 'o', 'o' , 'o', '^']
colors =['blue','blue', 'red', 'green', 'green']
titles = [subName + ' - Run 1 - Act Learning', 
          subName + ' - Run 1 - Stim Learning' ,
          subName + ' - Run 2 - Stim Learning',
          subName + ' - Run 2 - Act Learning']

for idx in range(columns*rows):
    fig.add_subplot(rows, columns, idx + 1)
    #create a list of y coordinates for every x coordinate
    for i in range(len(dfPlot)):
        plt.scatter(dfPlot.choise[i],[y[i] for j in range(len(dfPlot.choise[i]))], 
                    s=10, c=colors[i], marker=markers[i])


    plt.yticks(y,[]) #show the empy y axis label
    plt.legend(dfPlot.label, fontsize=8)
    plt.xlabel('Trials', fontsize=10)
    plt.title(titles[idx], fontsize=8)
    
plt.subplots_adjust(wspace=10., hspace = 10.)
plt.show()



# fig.savefig('out.png', dpi=100)
```
