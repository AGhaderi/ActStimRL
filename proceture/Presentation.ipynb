```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.figure(figsize=(8,4), tight_layout=True)
x5 = np.concatenate([np.arange(1, 101), 
                     np.arange(101, 201)]).reshape(-1,1)
y5 = np.concatenate([np.random.binomial(1, .9, 100), 
                     np.random.binomial(1, .1, 100)])

x6 = np.concatenate([np.arange(1, 121), 
                     np.arange(121, 201)]).reshape(-1,1)
y6 = np.concatenate([np.random.binomial(1, .8, 120),
                     np.random.binomial(1, .2, 80),])


x7 = np.concatenate([np.arange(1, 71), 
                     np.arange(71, 201)]).reshape(-1,1)
y7 = np.concatenate([np.random.binomial(1, .9, 130), 
                     np.random.binomial(1, .1, 70)])

model5 = LogisticRegression(solver='liblinear', random_state=0)
model5.fit(x5, y5)

model6 = LogisticRegression(solver='liblinear', random_state=0)
model6.fit(x6, y6)

model7 = LogisticRegression(solver='liblinear', random_state=0)
model7.fit(x7, y7)

plt.plot(np.unique(x5), model5.predict_proba(np.unique(x5).reshape(-1,1))[:,1])
plt.plot(np.unique(x7), model7.predict_proba(np.unique(x7).reshape(-1,1))[:,1])
plt.axhline(y=.5, color='black' , linestyle='--')
plt.axvline(x = 100, color='#ff7f0e', linewidth=1, alpha=.7)
plt.scatter(np.concatenate([x5[y5==1], x5[y5==0]]), np.concatenate([y5[y5==1]+.05, y5[y5==0]-.05]), s=10, c='#2ca02c', marker='o')
plt.xlabel('Trial', fontsize='14')
plt.ylabel('Probability of choice', fontsize='12')
plt.xticks([]) 
plt.savefig('stable_logistic_choice.png' , dpi=300)
```

```python
fig = plt.figure(figsize=(8,4), tight_layout=True)
x5 = np.concatenate([np.arange(1, 51), 
                     np.arange(51, 101), 
                     np.arange(101, 151)]).reshape(-1,1)
y5 = np.concatenate([np.random.binomial(1, .9, 50),
                     np.random.binomial(1, .1, 50), 
                     np.random.binomial(1, .9, 50)])

model5 = LogisticRegression(solver='liblinear', random_state=0)
model5.fit(x5[0:100], y5[0:100])

model6 = LogisticRegression(solver='liblinear', random_state=0)
model6.fit(x5[50:150], y5[50:150])


plt.plot(np.unique(x5[0:100]), model5.predict_proba(np.unique(x5[0:100]).reshape(-1,1))[:,1])
plt.plot(np.unique(x5[50:150]), model6.predict_proba(np.unique(x5[50:150]).reshape(-1,1))[:,1])
plt.axhline(y=.5, color='black' , linestyle='--')
plt.axvline(x = 50, color='#ff7f0e', linewidth=1, alpha=.7)
plt.axvline(x = 100, color='#ff7f0e', linewidth=1, alpha=.7)
plt.scatter(np.concatenate([x5[y5==1], x5[y5==0]]), np.concatenate([y5[y5==1]+.05, y5[y5==0]-.05]), s=10, c='#2ca02c', marker='o')
plt.xlabel('Trial', fontsize='14')
plt.ylabel('Probability of choice', fontsize='12')
plt.xticks([]) 
plt.savefig('volatile_logistic_choice.png' , dpi=300)
```
