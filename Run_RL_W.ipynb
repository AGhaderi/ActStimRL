```python
import numpy as np
import pandas as pd
import stan
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az
```

```python
# reading the RL model span
file_name = 'RL_W.stan' 
file_read = open(file_name, 'r')
stan_model = file_read.read()

# reading the observed data
examples_dir = "../data/sub-X99071_ses-01_task-DA_run-2_beh_20210422_14-27-10.csv"
df = pd.read_csv(examples_dir)
df_act = df[df.block == 'Act']


#set inistial data for molde span
data_init = {'N':int(df_act.shape[0]),     
             'pushed':np.array(df_act.pushed).astype(int),  
             'yellowChosen':np.array(df_act.yellowChosen).astype(int), 
             'winAmtPushable':np.array(df_act.winAmtPushable).astype(int), 
             'winAmtYellow':np.array(df_act.winAmtYellow).astype(int), 
             'rewarded':np.array(df_act.correctChoice).astype(int), 
             'p_push_init':.1, 
             'p_yell_init':.1}
```

```python
posterior = stan.build(stan_model, data = data_init)
fit = posterior.sample(num_chains=1000, num_samples=100)
```
