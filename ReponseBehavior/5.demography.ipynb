```python
############################################ Find Group, X number and age for inclusion_exclusion_subjects excel from subject_registry_Master excel file  
# 
#  read subject registry master
Data_Status = pd.read_excel('/mnt/projects/7TPD/bids/derivatives/fMRI_DA/AllBehData/subject_registry_Master.xlsx', sheet_name='Data_Status')

# read all avialable subjects
inclusion_exclusion_subjects = pd.read_excel('/mnt/projects/7TPD/bids/derivatives/fMRI_DA/AllBehData/inclusion_exclusion_subjects.xlsx')

# list of participants
Sub_ID = inclusion_exclusion_subjects['Sub-ID']
# get the group, x-number-age
for sub in Sub_ID:
    # there is no information recored for sub-020
    if sub.lower()!='sub-020':
        # find the relavant features in registry 
        extract_feature = Data_Status[Data_Status['Unnamed: 5'].str.lower()==sub.lower()][['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4']].to_numpy()[0]
        # put them into inlusion and exclusion 
        inclusion_exclusion_subjects.loc[inclusion_exclusion_subjects['Sub-ID']==sub, ['group', 'age', 'X-number']] = extract_feature

# save age and group disease
inclusion_exclusion_subjects.to_excel('/mnt/projects/7TPD/bids/derivatives/fMRI_DA/AllBehData/inclusion_exclusion_subjects1.xlsx')
```
